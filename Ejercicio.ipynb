### Conteo de valores nulos o vacios

```python
### conteo de valores null por columna
conteo_nulos = df.isnull().sum()
print(conteo_nulos)
```

### Valores almacenados por columnas (int, object, float...)

```python
# Al iterar un df se obtienen los nombres de las columnas 
for columna in df:
    #una forma de acceder a las columnas es usarlo como dic
    datos_columna = df[columna]
    # Cantidad de valores unicos
    distintos = datos_columna.nunique()

    print(f"La columna {columna} tienes {distintos} valores diferentes.")
```

### Eliminacion de Columnas irrelevantes en el data

#### Eliminacion de columnas con al menos un valor nulo

```python
df_sin_nulos = df.dropna(axis=1)
```

#### Eliminacion de columnas que tienen todos los valores vacios

```python
df_sin_nulos_totales = df.dropna(axis=1, how='all')
```

```python
conteo_nulos = df_sin_nulos_totales.isnull().sum()
print(conteo_nulos)
```

#### Conteo de valores nulos por columna

```python
conteo_nulos = df_sin_nulos.isnull().sum()
print(conteo_nulos)
```

#### Eliminar Columnas por alto % de datos nulos 

```python
limite = len(df)* 0.7
df_sin_muchos_nulos = df.dropna(axis=1, thresh=limite)
```

```python
conteo_nulos = df_sin_muchos_nulos.isnull().sum()
print(conteo_nulos)
```

#### Se selecciona el mejor metodo para eliminacion en este caso

```python
df = df_sin_muchos_nulos
```

```python
### Revisamos la inf de la data frame para visualizar cambios

print(f"Cantidad de filas: {len(df)}")
print(f"Dimensiones: {df.shape}")
print(f"Cantidad de valores: {df.size}")
print("Columnas: \n", *df.columns + '\n')
```
